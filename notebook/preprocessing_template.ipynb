# ðŸ”§ Data Preprocessing

```python
# 1. Import library
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
# 2. Load dataset langsung dari GitHub
df = pd.read_csv('https://github.com/nekowawolf/NASA-promise-dataset/raw/refs/heads/main/cm1.csv')
```

```python
# 3. Hapus missing value (drop rows)
df = df.dropna()
print("Ukuran data setelah drop NA:", df.shape)
```

```python
# 4. Pastikan kolom target bertipe numerik
df['defects'] = df['defects'].astype(int)
```

```python
# 5. Pisahkan fitur dan target
X = df.drop(columns='defects')
y = df['defects']
```

```python
# 6. Normalisasi fitur numerik
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# 7. Gabungkan kembali ke DataFrame
df_final = pd.DataFrame(X_scaled, columns=X.columns)
df_final['defects'] = y.values
```

```python
# 8. Simpan hasil preprocessing ke file
df_final.to_csv('cm1_final.csv', index=False)
print("\nDataset final disimpan sebagai 'cm1_final.csv'")
```
