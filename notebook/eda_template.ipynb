```python
# 1. Import library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 2. Load dataset langsung dari GitHub
df = pd.read_csv('https://github.com/nekowawolf/NASA-promise-dataset/raw/refs/heads/main/cm1.csv')
print("Ukuran dataset:", df.shape)
df.head()
```

```python
# 3. Info umum dan statistik deskriptif
df.info()
df.describe()
```

```python
# 4. Cek missing value
print("Missing values per kolom:")
print(df.isnull().sum())
```

```python
# 5. Visualisasi distribusi label target
sns.countplot(x='defects', data=df)
plt.title("Distribusi Label Defects")
plt.show()
```

```python
# 6. Korelasi antar fitur (heatmap)
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), annot=False, cmap='coolwarm')
plt.title("Korelasi antar Fitur")
plt.show()
```
